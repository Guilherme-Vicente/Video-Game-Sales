```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
#Leitura do Arquivo :
df = pd.read_csv('/Users/guilherme/Downloads/vgsales.csv')
```

Fazendo analise de alguns dados do dataset encontrado no kaggle,e aproveitando para fazer uma pequena analise dos jogos que mais joguei, que no caso seria o Call of duty black ops 2 e Call of duty modern warfare 4.

## Referente aos Dados :

1. Rank - Ranking
2. Name - Nome
3. Platform - Plataforma
4. Year - Ano
5. Genre - Gênero
6. Publisher - Editora
7. NA_Sales - Vendas América do Nort
8. EU_Sales - Vendas Europa
9. JP_Sales - Vendas Japão
10. Other_Sales - Outras vendas
11. Global_Sales - Vendas global

```python
# Ao invés de apagar linhas ou colunas que estao com valor nulo(NaN),substitui por 'Desconhecido'.
df.update(df['Year'].fillna('Desconhecido'))
df.update(df['Publisher'].fillna('Desconhecido'))
```

```python
df.shape
```

```python
# Quantidade de jogos lancados para cada plataforma.
df['Platform'].value_counts().sort_values()
```

```python
maisvendidos = df[['Name', 'Global_Sales']].head(15).set_index('Name').sort_values('Global_Sales')
```

```python
# Os 15 Jogos mais vendidos.
maisvendidos
```

# Grafico dos 15 jogos mais vendidos.

```python
maisvendidos.plot(kind='barh', figsize=(11,7), grid=False, color='green', legend=False)

plt.title('Os 15 jogos mais vendidos no mundo')
plt.xlabel('Total de vebdas (em milhões de doláres)')
plt.show()
```
